```python
import json

# 1. Daftar provinsi yang ingin diambil
provinsi_yang_diinginkan = {
    'Papua Selatan', 
    'Papua Tengah', 
    'Papua Barat Daya', 
    'Papua Pegunungan'
}

file_sumber_papua = 'Batas_Provinsi_Indonesia.geojson'
file_target_gadm = 'gadm41_IDN_1.json'
file_hasil_gabungan = 'gadm41_IDN_1_DENGAN_PAPUA.json'

fitur_papua_baru = []

print("Membaca file sumber (Batas_Provinsi_Indonesia.geojson)...")
# 2. Buka file sumber dan filter provinsi yang diinginkan
try:
    with open(file_sumber_papua, 'r', encoding='utf-8') as f:
        data_sumber = json.load(f)

    for feature in data_sumber['features']:
        # GANTI 'WADMPR' jika nama properti di file Anda berbeda!
        nama_provinsi = feature['properties'].get('WADMPR')
        
        if nama_provinsi in provinsi_yang_diinginkan:
            fitur_papua_baru.append(feature)
            print(f" -> Ditemukan: {nama_provinsi}")

    if not fitur_papua_baru:
        print("PERINGATAN: Tidak ada provinsi yang cocok ditemukan di file sumber.")
        print("Periksa kembali nama kunci ('WADMPR') dan ejaan di 'provinsi_yang_diinginkan'.")

except FileNotFoundError:
    print(f"Error: File {file_sumber_papua} tidak ditemukan.")
    exit()
except Exception as e:
    print(f"Error saat membaca file sumber: {e}")
    exit()


print(f"\nMembaca file target ({file_target_gadm})...")
# 3. Buka file target dan tambahkan fitur-fitur baru
try:
    with open(file_target_gadm, 'r', encoding='utf-8') as f:
        data_target = json.load(f)

    # 4. Tambahkan list fitur papua baru ke list fitur yang sudah ada
    data_target['features'].extend(fitur_papua_baru)
    
    print(f"Berhasil menambahkan {len(fitur_papua_baru)} provinsi baru ke file target.")

except FileNotFoundError:
    print(f"Error: File {file_target_gadm} tidak ditemukan.")
    exit()
except Exception as e:
    print(f"Error saat membaca file target: {e}")
    exit()

# 5. Simpan hasilnya ke file baru
try:
    with open(file_hasil_gabungan, 'w', encoding='utf-8') as f:
        json.dump(data_target, f, indent=2) # indent=2 agar file rapi
    
    print(f"\nBerhasil! File gabungan disimpan sebagai: {file_hasil_gabungan}")
    print(f"Total provinsi di file baru: {len(data_target['features'])}")

except Exception as e:
    print(f"Error saat menyimpan file baru: {e}")
```
